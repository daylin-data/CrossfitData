### Step 1: Import Libraries

```python
import pandas as pd # for data manipulation
import numpy as np  # for handling missing values
```

### Step 2: Load the Dataset

```python
# Load the dataset
file_path = "df_games_and_open.csv"
data = pd.read_csv(file_path)

# Display first few rows
data
```

### Step 3: Inspect the Dataset

```python
# Get database infomation
print("\nDataset Info:")
print(data.info())
```

```python
# Check for missing numbers
print(data.isnull().sum())
```

### Step 4: Handle Missing Values

```python
# If needed drop rows with missing values
data = data.dropna(subset=["lastName"])

# Fill missing values in less critical columns
data["affiliateName"] = data["affiliateName"].fillna("Unknown")
data["countryOfOriginName"] = data["countryOfOriginName"].fillna("Unknown")
data["regionName"] = data["regionName"].fillna("Unknown")

print(data.info())
```

### Step 5: Remove Duplicates

```python
# Check for duplicates
duplicates = data.duplicated().sum()
print(f"\nNumber of Duplicates: {duplicates}")

# Drop duplicates
data = data.drop_duplicates()

#Verify duplicates are removed
print(f"\nNumber of Duplicates Remaining: {data.duplicated().sum()}")
```

### Step 6: Fix Data Types

```python
# Convert quantitative data to numberic
data["height"] = pd.to_numeric(data["height"], errors="coerce")

print(data["height"])
```

### Step 7: Detect and Handle Outliers

##### Calculate IQR 
q1 = data[""].quantile(.25)
q3 = data[""].quantile(.75)

iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

##### Filter out outliers 
data = data[(data[""] >= lower_bound) & (data[""] <= upper_bound)]
