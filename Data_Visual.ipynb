## Visualizing the Data

#### Get the tools

```python
import pandas as pd # for data manipulation
import numpy as np  # for handling missing values
import plotly
```

#### Import Data

```python
# Load the dataset
file_path = "df_games_and_open.csv"
data = pd.read_csv(file_path)

# Display first few rows
data
```

## See the Data

```python
import plotly.express as px

# Create a histrogram of overallScore
fig = px.histogram(
    data, 
    x="overallScore",  
    title="Overall Score Distribution", 
    labels={"Overall Score": "overallScore"}, 
    color_discrete_sequence=["#636EFA"]

)

# Update Layout for a vibrant appearance
fig.update_layout(
    xaxis_title= "Overall Score",
    yaxis_title= "Frequency", 
    title_font=dict(size=20, 
    color="black"), 
    template="gridon"

)

# Show the plot
fig. show()
```
